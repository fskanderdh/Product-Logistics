```python
from google.colab import auth
auth.authenticate_user()
from google.cloud import bigquery
from google.cloud.bigquery import magics
%load_ext google.cloud.bigquery
```

```python
client = bigquery.Client(project='logistics-rider-staging')
project_id = 'logistics-rider-staging'
magics.context.project = 'logistics-rider-staging'
```

```python
df = client.query('''
SELECT
  EXTRACT(HOUR FROM DATETIME(delivery_created_at, "UTC") + INTERVAL 9 HOUR) AS hour_of_day_kst,
  (SUM(rider_accepted_count) - sum(manual_undispatched_count))/ SUM(rider_notified_count) AS actual_acceptance_ratio
from fulfillment-dwh-production.cl.acceptance_rate_deliveries
where country_code='kr2'
and created_date_local IN ('2024-09-12','2024-09-11','2024-09-10','2024-09-09')
GROUP BY hour_of_day_kst
ORDER BY hour_of_day_kst;
''').to_dataframe()

# Display the resulting DataFrame
df.head(25)
```

```python
# prompt: Modify the query in the previous cell to include all weekdays in the past month

df2 = client.query('''
SELECT
  EXTRACT(HOUR FROM DATETIME(delivery_created_at, "UTC") + INTERVAL 9 HOUR) AS hour_of_day_kst,
  (SUM(rider_accepted_count) - sum(manual_undispatched_count))/ SUM(rider_notified_count) AS actual_acceptance_ratio
from fulfillment-dwh-production.cl.acceptance_rate_deliveries
where country_code='kr2'
and DATE(created_date_local) >= DATE_SUB(CURRENT_DATE(), INTERVAL 14 DAY)
and EXTRACT(DAYOFWEEK FROM created_date_local) BETWEEN 2 AND 6 -- Monday to Friday
GROUP BY hour_of_day_kst
ORDER BY hour_of_day_kst;
''').to_dataframe()

# Display the resulting DataFrame
df2.head(25)
```

```python
# prompt: Change the previous cell to include the weekends only

df3 = client.query('''
SELECT
  EXTRACT(HOUR FROM DATETIME(delivery_created_at, "UTC") + INTERVAL 9 HOUR) AS hour_of_day_kst,
  (SUM(rider_accepted_count) - sum(manual_undispatched_count))/ SUM(rider_notified_count) AS actual_acceptance_ratio
from fulfillment-dwh-production.cl.acceptance_rate_deliveries
where country_code='kr2'
and DATE(created_date_local) >= DATE_SUB(CURRENT_DATE(), INTERVAL 14 DAY)
and EXTRACT(DAYOFWEEK FROM created_date_local) IN (1, 7) -- Saturday and Sunday
GROUP BY hour_of_day_kst
ORDER BY hour_of_day_kst;
''').to_dataframe()

# Display the resulting DataFrame
df3.head(25)
```

```python
# prompt: Generate a bar chart plot for df2 and df3 showing actual acceptance ratio by hour of day (KST). Display values on the graph for each bar. Display a graph for df2 and df3 separately.

import matplotlib.pyplot as plt

# Plot for df2 (Weekdays)
plt.figure(figsize=(12, 6))
plt.bar(df2['hour_of_day_kst'], df2['actual_acceptance_ratio'])
plt.xlabel('Hour of Day (KST)')
plt.ylabel('Actual Acceptance Ratio')
plt.title('Actual Acceptance Ratio by Hour of Day (Weekdays)')

# Add value annotations to the bars
for i, v in enumerate(df2['actual_acceptance_ratio']):
  plt.text(df2['hour_of_day_kst'][i], v, str(round(v, 2)), ha='center', va='bottom')

# Set x-axis ticks with an interval of 1
plt.xticks(range(0, 24, 1))  # Assuming 'hour_of_day_kst' ranges from 0 to 23

plt.show()


# Plot for df3 (Weekends)
plt.figure(figsize=(12, 6))
plt.bar(df3['hour_of_day_kst'], df3['actual_acceptance_ratio'])
plt.xlabel('Hour of Day (KST)')
plt.ylabel('Actual Acceptance Ratio')
plt.title('Actual Acceptance Ratio by Hour of Day (Weekends)')

# Add value annotations to the bars
for i, v in enumerate(df3['actual_acceptance_ratio']):
  plt.text(df3['hour_of_day_kst'][i], v, str(round(v, 2)), ha='center', va='bottom')

# Set x-axis ticks with an interval of 1
plt.xticks(range(0, 24, 1))

plt.show()
```

```python
df_orders_we = client.query('''
SELECT
  EXTRACT(HOUR FROM DATETIME(delivery_created_at, "UTC") + INTERVAL 9 HOUR) AS hour_of_day_kst,
  COUNT(DISTINCT order_id)
from fulfillment-dwh-production.cl.acceptance_rate_deliveries
where country_code='kr2'
and DATE(created_date_local) >= DATE_SUB(CURRENT_DATE(), INTERVAL 14 DAY)
and EXTRACT(DAYOFWEEK FROM created_date_local) IN (1, 7) -- Saturday and Sunday
GROUP BY hour_of_day_kst
ORDER BY hour_of_day_kst;
''').to_dataframe()

# Display the resulting DataFrame
df_orders_we.head(25)
```

```python
df_orders = client.query('''
SELECT
  EXTRACT(HOUR FROM DATETIME(delivery_created_at, "UTC") + INTERVAL 9 HOUR) AS hour_of_day_kst,
  COUNT(DISTINCT order_id)
from fulfillment-dwh-production.cl.acceptance_rate_deliveries
where country_code='kr2'
and DATE(created_date_local) >= DATE_SUB(CURRENT_DATE(), INTERVAL 14 DAY)
and EXTRACT(DAYOFWEEK FROM created_date_local) BETWEEN 2 AND 6 -- Monday to Friday
GROUP BY hour_of_day_kst
ORDER BY hour_of_day_kst;
''').to_dataframe()

# Display the resulting DataFrame
df_orders.head(25)
```

```python
# prompt: Generate a bar chart with df_orders_we and df_orders per hour, on the same chart. Colors steelblue for df_orders and coral for df_orders_we

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot for df_orders (Weekdays)
plt.bar(df_orders['hour_of_day_kst'], df_orders['f0_'], color='steelblue', label='Weekdays', width=0.4)

# Plot for df_orders_we (Weekends)
plt.bar(df_orders_we['hour_of_day_kst'] + 0.4, df_orders_we['f0_'], color='coral', label='Weekends', width=0.4)

plt.xlabel('Hour of Day (KST)')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Hour of Day (Weekdays vs. Weekends)')

# Add value annotations to the bars for weekdays
for i, v in enumerate(df_orders['f0_']):
  plt.text(df_orders['hour_of_day_kst'][i], v, str(v), ha='center', va='bottom', fontsize=6)

# Add value annotations to the bars for weekends
for i, v in enumerate(df_orders_we['f0_']):
  plt.text(df_orders_we['hour_of_day_kst'][i] + 0.4, v, str(v), ha='center', va='bottom', fontsize=6)

# Set x-axis ticks with an interval of 1
plt.xticks(range(0, 24, 1))

plt.legend()
plt.show()
```

```python
# prompt: Display a combo chart with df2 and df3 as bars, and df_orders and df_orders_we as lines, on the same chart, but with two different Y axis.

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for df2 (Weekdays) and df3 (Weekends)
ax1.bar(df2['hour_of_day_kst'], df2['actual_acceptance_ratio'], color='steelblue', label='Weekdays Acceptance Ratio', width=0.4)
ax1.bar(df3['hour_of_day_kst'] + 0.4, df3['actual_acceptance_ratio'], color='coral', label='Weekends Acceptance Ratio', width=0.4)

ax1.set_xlabel('Hour of Day (KST)')
ax1.set_ylabel('Actual Acceptance Ratio')
ax1.set_title('Actual Acceptance Ratio and Number of Orders by Hour of Day (Weekdays vs. Weekends)')

# Add value annotations to the bars for weekdays
for i, v in enumerate(df2['actual_acceptance_ratio']):
  ax1.text(df2['hour_of_day_kst'][i], v, str(round(v, 2)), ha='center', va='bottom', fontsize=6)

# Add value annotations to the bars for weekends
for i, v in enumerate(df3['actual_acceptance_ratio']):
  ax1.text(df3['hour_of_day_kst'][i] + 0.4, v, str(round(v, 2)), ha='center', va='bottom', fontsize=6)

# Create a second y-axis for the line chart
ax2 = ax1.twinx()

# Line chart for df_orders (Weekdays) and df_orders_we (Weekends)
ax2.plot(df_orders['hour_of_day_kst'], df_orders['f0_'], color='darkblue', label='Weekdays Orders')
ax2.plot(df_orders_we['hour_of_day_kst'], df_orders_we['f0_'], color='orangered', label='Weekends Orders')

ax2.set_ylabel('Number of Orders')

# Set x-axis ticks with an interval of 1
plt.xticks(range(0, 24, 1))

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```
